```python
import csv
```

```python
streets = {}
coordinate_bus={}
```

```python
with open('bus_stops.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        if row['Street'] in streets:
            streets[ row['Street']]+=1
        else:
            streets[ row['Street']]=1
            coordinate_bus[ row['Street']]=row["Longitude_WGS84"],row["Latitude_WGS84"]          
          
      
```
